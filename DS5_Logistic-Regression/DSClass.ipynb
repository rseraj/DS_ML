```python
gender_encode = lambda x: 1 if x == 'male' else 0
df.Sex = df.Sex.apply(gender_encode)
df.head()
```

```python
df.Cabin.unique()
```

```python
def cabin_encode(cabin):

  try:

      if type(cabin) == float and np.isnan(cabin):

          return 0

      else:

          temp_cabin_list = cabin.split(" ")
          cab_sum = 0
          for cab in temp_cabin_list:

              letter = list(cab).pop(0)
              cab_sum += float(str(ord(letter)) + "." + cab[1:])

          return cab_sum

  except Exception as e:

      print(e)
      print(f"Faulty entry: {cabin}")
```

```python
df.Cabin = df.Cabin.apply(cabin_encode)
df.head()
```

```python
from multiprocessing import Pool
```

```python
def age_decade(age):

  return int(age // 10)

pool_dec = Pool().map(age_decade, df.Age)
df["Age_Decade"] = pool_dec
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.boxplot(data= tips, x= "day", y= "total_bill", palette= "coolwarm", hue= "day")
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
